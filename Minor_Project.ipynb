```
from bing_image_downloader import downloader
#downloading car images
downloader.download('car', limit=15,  output_dir='images', adult_filter_off=True, force_replace=False, timeout=60)
```

```
#downloading motorcycle images
downloader.download('motorcycle', limit=15,  output_dir='images', adult_filter_off=True, force_replace=False, timeout=60)
```

```
import numpy as np 
from skimage.transform import resize
from skimage.io import imread
import matplotlib.pyplot as plt
import os


target = [] #stores class labels  
images = [] #stores images pixel values
flat_data = [] #stores flattened image vector data
DATADIR = '/content/images'
CATEGORIES = ['car','motorcycle']
for category in CATEGORIES:
  class_num = CATEGORIES.index(category)
  path = os.path.join(DATADIR, category)
  for img in os.listdir(path):
    img_array = imread(os.path.join(path,img))
    img_resized = resize(img_array,(150,150,3))
    flat_data.append(img_resized.flatten())
    images.append(img_resized)
    target.append(class_num)
#converting to numpy arrays
target = np.array(target) 
images = np.array(images)
flat_data = np.array(flat_data)
```

```
import pandas as pd
#Creating DataFrame with flattened data
df = pd.DataFrame(flat_data)
df['Target'] = target
df
```

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
#Splitting data into training and test data
X = df.drop('Target',axis=1)
y = df['Target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```
#scaling data for more accurate predictions
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```
#find best value of n using GridSearchCV
from sklearn.model_selection import GridSearchCV
n = np.arange(1,24)
param = {'n_neighbors': n}
model = KNeighborsClassifier()
model_grid = GridSearchCV(model,param)
model_grid.fit(X,y)
model_grid.best_params_
```

```
#model trained taking K value = 7
model = KNeighborsClassifier(n_neighbors = 7, metric = 'euclidean')
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```
#Predictions by model
y_pred = model.predict(X_test)
print('Prediction :',y_pred)
#actual values
print('Actual Values :',np.array(y_test))
```

```
#accuracy score
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,np.array(y_test)))
```

```
#average accuracy of model found using cross validation
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, X, y, cv = 5)
np.mean(scores)
```

```
#Evaluating model using Confusion matrix and Classification Report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
print(confusion_matrix(np.array(y_test), y_pred))
# Printing the precision and recall, among other metrics
print(classification_report(np.array(y_test), y_pred))
```
